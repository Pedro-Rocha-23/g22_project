```python
import pandas as pd
```

```python
df = pd.read_csv("g22_Research_Grants.csv", sep = ';')
```

```python
df_university = df.loc[:,['university_id','name','creation_date']]
```

```python
df_university = df_university.drop_duplicates(subset='university_id', keep='first')
```

```python
df_grant = df.loc[:,['grant_id','title','category','director_id']]
```

```python
df_grant = df_grant.drop_duplicates(subset='grant_id', keep='first')
```

```python
df_lab = df.loc[:,['lab_id','extra_info','university_id']]
```

```python
df_lab = df_lab.drop_duplicates(subset='lab_id', keep='first')
```

```python
df_lab = df_lab.dropna()
```

```python
df_lab['lab_id'] = df_lab['lab_id'].astype(int)
```

```python
df_director = df.loc[:,['director_id','director_name','date_of_birth']]
```

```python
df_director = df_director.drop_duplicates(subset='director_id', keep='first')
```

```python
df_uni_grant = df.loc[:,['university_id','grant_id','amount','start_date']]
```

```python
df_uni_grant = df_uni_grant.drop_duplicates(subset=['university_id', 'grant_id'], keep='first')
```

```python
df_uni_grant.insert(0, 'id', range(1, len(df_uni_grant) + 1))
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///g22_db.db')
```

```python
df_university.to_sql('University', con=engine, if_exists='replace', index=False)
```

```python
df_grant.to_sql('Grant', con=engine, if_exists='replace', index=False)
```

```python
df_lab.to_sql('Lab', con=engine, if_exists='replace', index=False)
```

```python
df_director.to_sql('Director', con=engine, if_exists='replace', index=False)
```

```python
df_uni_grant.to_sql('Uni_grant', con=engine, if_exists='replace', index=False)
```

```python
engine.dispose()
```
